```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'E:\Timeseries_Upgrad\Metrail+dataset.csv')
data.columns = ['Month','Travellers']
```

```python
data
```

```python
data['Month'] = pd.to_datetime(data['Month'], format ='%b-%y')
data=data.set_index('Month')
```

```python
data
```

```python
data.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Travellers Data')
plt.show(block=False)
```

```python
data1=data.assign(Travellers_farward_imputation=data['Travellers'].fillna(method='ffill'))

data1[['Travellers_farward_imputation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Imputation with ffill')
plt.show(block=False)
```

```python
data1['Travellers'].mean()
```

```python
data2=data.assign(Travellers_farward_imputation=data['Travellers'].fillna(method='backfill'))

data2[['Travellers_farward_imputation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Imputation with ffill')
plt.show(block=False)
```

```python
data2['Travellers'].mean()
```

```python
data3=data.assign(Travellers_farward_imputation=data['Travellers'].fillna(method='pad'))

data3[['Travellers_farward_imputation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Imputation with ffill')
plt.show(block=False)
```

```python
data3.Travellers.mean()
```

# Linear Interpolation

```python
data=data.assign(Travellers_Linear_Interpolation=data.Travellers.interpolate(method='linear'))

data[['Travellers_Linear_Interpolation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Treatment with Linear Interpolation')
plt.show(block=False)
```

```python
data.Travellers.mean()
```
