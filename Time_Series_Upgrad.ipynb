# import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

# import time series data

```python
data = pd.read_csv(r'E:\Timeseries_Upgrad\airline-passenger-traffic.csv',header=None)
data.columns = ['Month','Passangers']
```

```python
data
```

```python
data['Month']=pd.to_datetime(data['Month'],format='%Y-%m')
data=data.set_index('Month')
```

```python
data
```

# Timeseries Analysis

```python
data.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline Passange Trafic Data')
plt.show(block=False)
```

# missing value treatment

# Mean Imputation:

```python
data=data.assign(Passangers_Mean_imputation=data.Passangers.fillna(data.Passangers.mean()))

data[['Passangers_Mean_imputation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Imputation with Mean')
plt.show(block=False)
```

# Linear Interpolation

```python
data=data.assign(Passangers_Linear_Interpolation=data.Passangers.interpolate(method='linear'))

data[['Passangers_Linear_Interpolation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing Value Treatment with Linear Interpolation')
plt.show(block=False)
```
